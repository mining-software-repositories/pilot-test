## 5. Verifica arquivos .java modificados em mais de um commit

### Dado um arquivo, procure se ele aparece em outros commits

[file1] -> [commit1, commit10]

[file2] -> [commit5]

[file3] -> [commit5, commit6]

[file4] -> [commit5, commit6, commit7]

...

[filen] -> [...]

### Carrega commits e arquivos em banco de dados

```python
for key, value in dicionario_arquivos_java_modificados.items():
    commit = dao.Commit(name=key)
    commitsCollection.insert_commit(commit)
    for each in value:
        my_file = dao.File(name=each, hash=key, commit_id=commit.id)
        filesCollections.insert_file(my_file)
```

```python
lista_de_arquivos_e_commits = filesCollections.query_all_files()

dicionario_files_com_commits = {}
lista_conjunto_arquivos = []
for each in lista_de_arquivos_e_commits:
    lista_conjunto_arquivos.append(each.name)

conjunto_arquivos = set(lista_conjunto_arquivos)

lista_temp = list(conjunto_arquivos)
lista_temp = sorted(lista_temp)

for each in lista_temp:
    lista_aux = []
    for item in lista_de_arquivos_e_commits:
        if each == item.name:
            lista_aux.append(item.hash)
    dicionario_files_com_commits[each] = lista_aux

for key, value in dicionario_files_com_commits.items():
    value = ','.join(value)
    print(f'{key}:{value}')
```

#### Gera arquivo com arquivo e seus commits

```python
file_name = 'arquivos-commmits-cassandra-3-1-ao-3-2.txt'
print(f'Criando o arquivo {file_name} ...')
linha = ''
with open(file_name, mode='w+', encoding='utf8') as f:
    for key, value in dicionario_files_com_commits.items():
        linha = key + ':' + ','.join(value)
        f.write(linha)
        f.write('\n')
        linha = ''
print(f'Arquivo {file_name} salvo com sucesso!')
```
