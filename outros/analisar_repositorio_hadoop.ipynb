```python
import entities
import utils
```

```python
my_commit = 'c5281a85e108c7cab6e84dfe351843483d720b9b'
repositorio_git_hadoop = '/Users/armandosoaressousa/git/msr/hadoop'
tag1 = 'rel/release-3.0.1'
tag2 = 'rel/release-3.1.0'
```

## 1. Lista de commits do Hadoop

```python
lista_commits_hadoop = utils.list_commits_hash_from_repository(repositorio_git_hadoop)
print(f'Total de commits: {len(lista_commits_hadoop)}')
```

## 2. Dado um commit mostra os arquivos modificados

```python
i = 1
for m in utils.list_modified_files_from_commit(my_commit, repositorio_git_hadoop):
    mf = entities.MyModifiedFile(m)
    print(f"{i} - Name: {mf.get_filename()}, change_type: {mf.get_change_type()}, Modifications: {mf.get_modifications()},  Number of lines: {mf.get_nloc()}, Complexity: {mf.get_complexity()}")
    i += 1
```

## 3. Lista commits entre duas tags

```python
lista_commits_entre_tags = utils.list_commits_between_tags(from_tag=tag1, to_tag=tag2, my_repository=repositorio_git_hadoop) 
print(len(lista_commits_entre_tags) )
```

## 4. Para cada commit entre as tags mostre os arquivos modificados

### Mostra todos os arquivos modificados entre as tags

```python
for each in lista_commits_entre_tags:
    print(each.hash)
    for m in utils.list_modified_files_from_commit(each.hash, repositorio_git_hadoop):
        mf = entities.MyModifiedFile(m)
        print(f"{i} - Name: {mf.get_filename()}, change_type: {mf.get_change_type()}, Modifications: {mf.get_modifications()},  Number of lines: {mf.get_nloc()}, Complexity: {mf.get_complexity()}")
        i += 1
```

### Mostra apenas os arquivos .java modificados entre as tags

```python
lista_conjunto_arquivos_java_modificados = []
lista_aux = []
for commit in lista_commits_entre_tags:
    print(commit.hash)
    i = 1
    for m in utils.list_modified_files_from_commit(commit.hash, repositorio_git_hadoop):
        mf = entities.MyModifiedFile(m)
        name = mf.get_filename()
        if '.java' in name:
            print(f"{i} - Name: {mf.get_filename()}, change_type: {mf.get_change_type()}, Modifications: {mf.get_modifications()},  Number of lines: {mf.get_nloc()}, Complexity: {mf.get_complexity()}")
            lista_aux.append(name)
            i += 1  
    i = 0  
    lista_aux.insert(0, commit.hash)
    lista_conjunto_arquivos_java_modificados.append(lista_aux)
    lista_aux = []
print('')
print(f'conjunto: {lista_conjunto_arquivos_java_modificados}')
```

```python
file_name = 'analise-3-03-1.txt'
print(f'Criando o arquivo {file_name} ...')
with open(file_name, mode='w+', encoding='utf8') as f:
    for conjunto in lista_conjunto_arquivos_java_modificados:
        for item in conjunto:
            linha = linha + item + ','
        linha = '[' + linha + ']'
        f.write(linha)
        f.write('\n')
        linha = ''
print(f'Arquivo {file_name} salvo com sucesso!')
```
